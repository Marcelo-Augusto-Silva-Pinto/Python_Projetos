```python
#Script para gerar Tabela Cliente

#Rascunho

import random
import pyodbc  # Biblioteca para conexão com o banco de dados SQL

# Listas dos 50 nomes e sobrenomes mais comuns no Brasil
nomes = [
    "Maria", "José", "Ana", "João", "Antônio", "Francisco", "Carlos", "Paulo", "Pedro", "Lucas",
    "Luiz", "Marcos", "Luís", "Rafael", "Gabriel", "Daniel", "Marcelo", "Bruno", "Eduardo", "Felipe",
    "Raimundo", "Miguel", "Arthur", "Gustavo", "Leonardo", "Thiago", "Matheus", "Enzo", "Guilherme", "Davi",
    "Larissa", "Amanda", "Juliana", "Patrícia", "Aline", "Camila", "Bruna", "Fernanda", "Isabela", "Mariana",
    "Tatiana", "Vanessa", "Carolina", "Bianca", "Clara", "Vitória", "Letícia", "Laura", "Beatriz", "Júlia"
]

sobrenomes = [
    "Silva", "Santos", "Oliveira", "Souza", "Rodrigues", "Ferreira", "Alves", "Pereira", "Lima", "Gomes",
    "Costa", "Ribeiro", "Martins", "Carvalho", "Araújo", "Melo", "Barbosa", "Monteiro", "Cardoso", "Teixeira",
    "Mendes", "Nascimento", "Cavalcanti", "Dias", "Castro", "Campos", "Bezerra", "Fernandes", "Vieira", "Correia",
    "Sousa", "Machado", "Rocha", "Neves", "Nunes", "Moreira", "Medeiros", "Freitas", "Barros", "Pinto",
    "Andrade", "Moura", "Dantas", "Cunha", "Brito", "Leal", "Farias", "Peixoto", "Fonseca", "Azevedo"
]

# Lista de nomes femininos
nomes_femininos = [
    "Maria", "Ana", "Larissa", "Amanda", "Juliana", "Patrícia", "Aline", "Camila", "Bruna", "Fernanda",
    "Isabela", "Mariana", "Tatiana", "Vanessa", "Carolina", "Bianca", "Clara", "Vitória", "Letícia", "Laura",
    "Beatriz", "Júlia"
]

provedores = ["gmail.com", "yahoo.com", "hotmail.com", "outlook.com"]

# Função para gerar um nome completo, gênero e e-mail
def gerar_nome_completo():
    nome = random.choice(nomes)
    sobrenome = random.choice(sobrenomes)
    genero = "F" if nome in nomes_femininos else "M"
    numero = random.randint(10, 99)
    usuario = f"{nome.lower()}{sobrenome.lower()}{numero}"
    provedor = random.choice(provedores)
    email = f"{usuario}@{provedor}"  # Formatar o e-mail

    return nome, sobrenome, genero, email

# Configurações de conexão com o banco de dados
server = "DESKTOP-HV85ADA\\SERVERSQL22"  # Substitua pelo nome do servidor
database = "TESTE"  # Substitua pelo nome do banco de dados
#username = "SEU_USUARIO"  # Substitua pelo seu usuário
#password = "SUA_SENHA"  # Substitua pela sua senha


#conn_str = f"DRIVER={{SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}" # String de conexão com user e senha

conn_str = f"DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;" # String de conexão pelo windows

# Conectar ao banco de dados
try:
    conexao = pyodbc.connect(conn_str)
    cursor = conexao.cursor()
    print("Conexão com o banco de dados estabelecida com sucesso!")
except Exception as e:
    print(f"Erro ao conectar ao banco de dados: {e}")
    exit()

# Gerar 1000 nomes completos com gênero e e-mail
dados_nomes = [gerar_nome_completo() for _ in range(1000000)]

# Inserir os dados na tabela 'cliente'
try:
    for nome, sobrenome, genero, email in dados_nomes:
        query = """
        INSERT INTO CLIENTE1 (NOME, SOBRENOME, GENERO, EMAIL) VALUES (?, ?, ?, ?)
        """
        cursor.execute(query, nome, sobrenome, genero, email)
    conexao.commit()  # Confirmar as inserções
    print("Dados inseridos na tabela 'cliente' com sucesso!")
except Exception as e:
    print(f"Erro ao inserir dados no banco de dados: {e}")
finally:
    
    cursor.close()
    conexao.close()
    print("Conexão com o banco de dados fechada.")
```
