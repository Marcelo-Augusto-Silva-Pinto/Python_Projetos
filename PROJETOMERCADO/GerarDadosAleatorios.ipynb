```python
#Script para gerar horarios aleatorios
import random
import pandas as pd

qtd_horarios = 1000

def gerar_horario():
    hora = random.randint(0,23)
    minuto = random.randint(0,59)
    hora_formatada = f"{hora:02d}:{minuto:02d}"
    return hora_formatada

lista_horarios = [gerar_horario() for _ in range(qtd_horarios)]

df_horarios= pd.DataFrame(lista_horarios, columns=["Horario"])

df_horarios.head(10)
```

```python
#Script para gerar datas aleatorias
import random
from datetime import date
import pandas as pd

def gerar_data(ano_inicio,ano_fim):
    ano = random.randint(ano_inicio, ano_fim)
    mes = random.randint(1, 12)
    if mes in [4, 6, 9, 11]:  # Meses com 30 dias
        dia = random.randint(1, 30)
    elif mes == 2:  # Fevereiro (considerando ano não bissexto)
        dia = random.randint(1, 28)
    else:  # Meses com 31 dias
        dia = random.randint(1, 31)

    #data_formatada = f"{ano:04d}/{mes:02d}/{dia:02d}" #data em formato string
    return date(ano, mes, dia)

#data_aleatoria = gerar_data(2003,2005)
#print(data_aleatoria)

lista_datas = [gerar_data(2021,2023) for _ in range(1000)]

df_datas = pd.DataFrame(lista_datas, columns=["Data"])

#Converte as datas para o formato br (em string)
#df_datas["Data"] = df_datas["Data"].apply(lambda x: x.strftime("%d/%m/%Y")) 

df_datas
```

```python
#Script para gerar datas aleatorias com horario

import random
from datetime import datetime
import pandas as pd

def gerar_data_hora(ano_inicio, ano_fim):
    ano = random.randint(ano_inicio, ano_fim)
    mes = random.randint(1, 12)
    if mes in [4, 6, 9, 11]:  
        dia = random.randint(1, 30)
    elif mes == 2:  
        dia = random.randint(1, 28)
    else:  
        dia = random.randint(1, 31)

    hora = random.randint(8, 22)
    minuto = random.randint(0, 59)
    segundo = random.randint(0, 59)
    
    return datetime(ano, mes, dia, hora, minuto, segundo)

# Gerar 10 datas e horas aleatórias
lista_datas_horas = [gerar_data_hora(2021, 2023) for _ in range(1000)]

df_datas_horas = pd.DataFrame(lista_datas_horas, columns=["Data_Hora"])

df_datas_horas.head(1000)
```
