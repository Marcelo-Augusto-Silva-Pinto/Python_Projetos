```python
#Script para gerar numero simplificado

import random
import pandas as pd

#Primeiro range define o tamanho do numero e o segundo define a quantidade de numeros gerados
lista_inteiros = [''.join([str(random.randint(0, 9)) for _ in range(10)]) for _ in range(1000)]

df_inteiro= pd.DataFrame(lista_inteiros)

df_inteiro.head(10)
```

```python
#Script para gerar numero decimal

import random
import pandas as pd

#num = int(input(print("digite um numero:")))
num = 1000
#Range define a quantidade de numeros gerados
lista_decimais = [round(random.uniform(2.5, 10), 2) for _ in range(num)]

df_decimal= pd.DataFrame(lista_decimais, columns=["Numero Decimal"])

df_decimal
```

```python
#Script para gerar horarios aleatorios
import random
import pandas as pd

qtd_horarios = 1000

def gerar_horario():
    hora = random.randint(0,23)
    minuto = random.randint(0,59)
    hora_formatada = f"{hora:02d}:{minuto:02d}"
    return hora_formatada

lista_horarios = [gerar_horario() for _ in range(qtd_horarios)]

df_horarios= pd.DataFrame(lista_horarios, columns=["Horario"])

df_horarios.head(10)
```

```python
#Script para gerar datas aleatorias
import random
from datetime import date
import pandas as pd

def gerar_data(ano_inicio,ano_fim):
    ano = random.randint(ano_inicio, ano_fim)
    mes = random.randint(1, 12)
    if mes in [4, 6, 9, 11]:  # Meses com 30 dias
        dia = random.randint(1, 30)
    elif mes == 2:  # Fevereiro (considerando ano não bissexto)
        dia = random.randint(1, 28)
    else:  # Meses com 31 dias
        dia = random.randint(1, 31)

    #data_formatada = f"{ano:04d}/{mes:02d}/{dia:02d}" #data em formato string
    return date(ano, mes, dia)

#data_aleatoria = gerar_data(2003,2005)
#print(data_aleatoria)

lista_datas = [gerar_data(2021,2023) for _ in range(1000)]

df_datas = pd.DataFrame(lista_datas, columns=["Data"])

#Converte as datas para o formato br (em string)
#df_datas["Data"] = df_datas["Data"].apply(lambda x: x.strftime("%d/%m/%Y")) 

df_datas
```

```python
#Script para gerar datas aleatorias com horario

import random
from datetime import datetime
import pandas as pd

def gerar_data_hora(ano_inicio, ano_fim):
    ano = random.randint(ano_inicio, ano_fim)
    mes = random.randint(1, 12)
    if mes in [4, 6, 9, 11]:  
        dia = random.randint(1, 30)
    elif mes == 2:  
        dia = random.randint(1, 28)
    else:  
        dia = random.randint(1, 31)

    hora = random.randint(8, 22)
    minuto = random.randint(0, 59)
    segundo = random.randint(0, 59)
    
    return datetime(ano, mes, dia, hora, minuto, segundo)

# Gerar 10 datas e horas aleatórias
lista_datas_horas = [gerar_data_hora(2021, 2023) for _ in range(1000)]

df_datas_horas = pd.DataFrame(lista_datas_horas, columns=["Data_Hora"])

df_datas_horas.head(1000)
```

```python
#Script para gerar nomes completos e o genero
import random

# Listas dos 50 nomes e sobrenomes mais comuns no Brasil
nomes = [
    "Maria", "Jose", "Ana", "Joao", "Antonio", "Francisco", "Carlos", "Paulo", "Pedro", "Lucas",
    "Luiz", "Marcos", "Luis", "Rafael", "Gabriel", "Daniel", "Marcelo", "Bruno", "Eduardo", "Felipe",
    "Raimundo", "Miguel", "Arthur", "Gustavo", "Leonardo", "Thiago", "Matheus", "Enzo", "Guilherme", "Davi",
    "Larissa", "Amanda", "Juliana", "Patricia", "Aline", "Camila", "Bruna", "Fernanda", "Isabela", "Mariana",
    "Tatiana", "Vanessa", "Carolina", "Bianca", "Clara", "Vitoria", "Leticia", "Laura", "Beatriz", "Julia"
]

sobrenomes = [
    "Silva", "Santos", "Oliveira", "Souza", "Rodrigues", "Ferreira", "Alves", "Pereira", "Lima", "Gomes",
    "Costa", "Ribeiro", "Martins", "Carvalho", "Araujo", "Melo", "Barbosa", "Monteiro", "Cardoso", "Teixeira",
    "Mendes", "Nascimento", "Cavalcanti", "Dias", "Castro", "Campos", "Bezerra", "Fernandes", "Vieira", "Correia",
    "Sousa", "Machado", "Rocha", "Neves", "Nunes", "Moreira", "Medeiros", "Freitas", "Barros", "Pinto",
    "Andrade", "Moura", "Dantas", "Cunha", "Brito", "Leal", "Farias", "Peixoto", "Fonseca", "Azevedo"
]

# Lista de nomes femininos
nomes_femininos = [
    "Maria", "Ana", "Larissa", "Amanda", "Juliana", "Patricia", "Aline", "Camila", "Bruna", "Fernanda",
    "Isabela", "Mariana", "Tatiana", "Vanessa", "Carolina", "Bianca", "Clara", "Vitoria", "Leticia", "Laura",
    "Beatriz", "Julia"
]

provedores = ["gmail.com", "yahoo.com", "hotmail.com", "outlook.com"]

# Função para gerar um nome completo e determinar o gênero
def gerar_nome_completo():
    nome = random.choice(nomes)
    sobrenome = random.choice(sobrenomes)
    genero = "Feminino" if nome in nomes_femininos else "Masculino"
    numero = random.randint(10, 99)
    usuario = f"{nome.lower()}{sobrenome.lower()}{numero}"    
    provedor = random.choice(provedores)       
    email = f"{usuario}@{provedor}" # Formatar o e-mail

    return nome, sobrenome, genero, email
 

# Gerar 100 nomes completos com gênero
dados_nomes = [gerar_nome_completo() for _ in range(1000)]

lista_nomes = [f"{nome} {sobrenome}" for nome, sobrenome, _, _ in dados_nomes]
lista_genero = [genero for _, _, genero, _ in dados_nomes]
lista_emails = [email for _, _, _, email in dados_nomes]

print("Lista de Nomes Completos:")
print(lista_nomes)

print("\nLista de Gêneros:")
print(lista_genero)

print("\nLista de E-mails:")
print(lista_emails[:10])
```
