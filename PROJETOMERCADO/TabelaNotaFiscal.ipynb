```python
#Script para gerar uma nota fiscal

import random
from datetime import datetime, date
import pandas as pd

quantidade_NFs = 100

def gerar_data_hora(ano_inicio, ano_fim):
    ano = random.randint(ano_inicio, ano_fim)
    mes = random.randint(1, 12)
    if mes in [4, 6, 9, 11]:  
        dia = random.randint(1, 30)
    elif mes == 2:  
        dia = random.randint(1, 28)
    else:  
        dia = random.randint(1, 31)

    hora = random.randint(8, 22)
    minuto = random.randint(0, 59)
    segundo = random.randint(0, 59)
    
    return datetime(ano, mes, dia, hora, minuto, segundo)

# Gerar datas e horas aleat√≥rias
lista_datas_horas = [gerar_data_hora(2021, 2023) for _ in range(quantidade_NFs)]

lista_ID_FORMA_PGMT = [random.randint(1,8) for _ in range(quantidade_NFs)]

'''
AJUSTAR A QUANTIDADE DE CLIENTES, NO FUTURO COLOCAR UMA VARIAVEL QUE ESTA CAPTURANDO A QUANTIDADE DE CLIENTE E 
SUBSTITUIR PELO HARD COD

'''
lista_ID_CLIENTE = [random.randint(1,1000) for _ in range(quantidade_NFs)]

lista_ID_LOJA = [random.randint(1,150) for _ in range(quantidade_NFs)]

lista_ID_CFOP = [random.randint(1,5) for _ in range(quantidade_NFs)]

df_notafiscal = pd.DataFrame({ 
                    "DATA":           lista_datas_horas,
                    "TOTAL_NOTA":     " ",
                    "ID_FORMA_PGMT":  lista_ID_FORMA_PGMT,
                    "ID_CLIENTE":     lista_ID_CLIENTE,
                    "ID_LOJA":        lista_ID_LOJA,
                    "ID_CFOP":        lista_ID_CFOP,
                    "DATA_CRIACAO":   date.today().strftime('%d/%m/%Y'),
                    "DATA_ATLZ":      "",
                    "VERSAO":         1,

})

df_notafiscal
```
